#### Task 3: Structural Reasoning (Simulation before Specification)
A PRD fails when it misses "Edge Cases" (e.g., What if the user has no internet? What if the fridge is empty?).
*   **Strategy:** Use **Chain-of-Thought** prompting.
*   **The Prompt:** Do *not* ask for the PRD yet. Instead, command the AI to:
    1.  `<simulation>`: Simulate a user trying to use the app in a **difficult** scenario (poor lighting, bad internet, unidentified vegetables). Write out the narrative of their struggle step-by-step.
    2.  `<requirements>`: Based *only* on that simulation, list the technical requirements needed to solve those specific struggles (e.g., "Offline Mode," "Low-Light Image Enhancement").
*   **Goal:** Force the model to "live" the problem before it solves it.

````python
# ==========================================
# TASK 3: STRUCTURAL REASONING (SIMULATION)
# ==========================================
print("\n--- TASK 3: The Edge Case Simulation (Chain of Thought) ---")

prompt_3 = f"""
We are building {product_idea}.
Do NOT write the PRD yet.

Step 1: <simulation>
Simulate a user named 'Tired Tony'. It is 8 PM, his kitchen lighting is dim,
he has spotty WiFi, and his fridge is messy with a half-eaten rotisserie chicken
and some wilted kale. Narrate his attempt to use the app and the frustrations he faces.
</simulation>

Step 2: <requirements>
Based EXCLUSIVELY on Tony's struggle, list 4 technical requirements to solve his pain points.
(e.g., if lighting is bad, what tech do we need? If wifi is bad, how do we handle image upload?)
</requirements>
"""

response_3 = get_completion(prompt_3)
print(response_3)

# OBSERVATION:
# The simulation forces the AI to "realize" that low-light enhancement and
# client-side compression (for bad wifi) are necessary features.
# A zero-shot prompt usually misses these edge cases.
````

## Part 1: Critical Analysis of Proposed Requirements

These requirements are architecturally naive and introduce significant vectors for security failures, hallucination escalation, and privacy breaches.

### Analysis of Requirement 1: Robust Low-Light Image Pre-Processing

**The Risk (My Fears):**
1.  **Hallucination Amplification:** Pre-processing images *locally* before cloud analysis can introduce **algorithmic artifacts**. If the sharpening filter overcompensates for the blur, the system might "see" phantom textures that look like perfectly good meat when it’s actually just fat residue or shadows. This forces the downstream CV model to hallucinate a high-confidence result based on corrupted input data.
2.  **Performance Overkill:** Trying to cram complex image enhancement (like full adaptive histogram equalization) onto older, low-end mobile hardware using local processing can introduce significant *new* latency, defeating the entire purpose of minimizing perceived lag.
3.  **Unnecessary Client Payload:** We are now shipping potentially large, complex local processing libraries to every device, increasing the app size and maintenance burden.

### Analysis of Requirement 2: Offline/Asynchronous Cloud Processing & Resume

**The Risk (My Fears):**
1.  **Data Inconsistency (The Stale State Nightmare):** If the upload fails, we save the *image metadata* locally. When the network recovers, we seamlessly resume. What if Tony opens the app 3 hours later? The image finally uploads, but in the interim, he *physically* threw out the chicken. The system will now present him with recipe options based on an image that is factually obsolete. **This is a recipe for user-facing data poisoning.**
2.  **Security Exposure:** Saving large image payloads locally, even temporarily, increases the device's attack surface. If the local storage isn't rigorously encrypted and access-controlled, we’ve just created a honeypot of private kitchen imagery.

### Analysis of Requirement 3: Granular Object Status Recognition (State Detection)

**The Risk (My Fears):**
1.  **Hallucination Catastrophe (The Spoiled Ingredient Problem):** This is the single biggest threat. The system is *required* to suggest recipes based on **"Wilted," "Leftover Bone," and "Uncertain"** inputs. If the CV model misinterprets "yellowing edges" as "slightly yellowed, still usable kale," and the recipe generator suggests using it in a *lightly cooked* dish, we have now leveraged customer trust to serve them potentially unsafe food recommendations. **We are liable for food safety suggestions based on shaky visual evidence.**
2.  **Model Complexity & Drift:** Training a model robustly enough to differentiate between "limp but safe" and "inedible slime" from a blurry photo is an AI moonshot. This complexity guarantees high false positive/negative rates.

### Analysis of Requirement 4: Rapid Local Re-Evaluation and Immediate Feedback Loop

**The Risk (My Fears):**
1.  **Network Dance for Simple Edits:** Forcing rapid client-side re-evaluation by minimizing network calls is fine for *display*, but if the core recipe database resides centrally, any significant change (like drastically reducing ingredient quantities) *still* requires a cloud call. If we try to lie to the user by calculating recipes locally on a massive, centralized database model, we'll fail spectacularly or, worse, expose proprietary logic locally.
2.  **Privacy Leakage on Edits:** If every granular edit triggers a micro-upload to update the "true" state of the fridge, we are tracking user behavior at an incredibly invasive level—not just what they *have*, but what they *admit* they have after realizing the AI was wrong.
